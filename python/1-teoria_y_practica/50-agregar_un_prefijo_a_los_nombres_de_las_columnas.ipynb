### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Agregar un prefijo a los nombres de las columnas

```
from pyspark.sql.functions import col, regexp_extract

data = [(1,'nombre A','Gibson'),
        (2,'nombre B','Fender'),
        (3,'nombre C','Ibanez')
       ]

df = spark.createDataFrame(data).toDF('id','nombre','empresa')
df.printSchema()
df.show(truncate=False)
```

```
prefix = 'sp_'
df_prefijo = df.select([col(c).alias(prefix+c) for c in df.columns])

display (df_prefijo)
```
