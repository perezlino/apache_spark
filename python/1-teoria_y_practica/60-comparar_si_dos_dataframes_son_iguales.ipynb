### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Comparar dos dataframes sin son iguales

Ambos tienen los mismos registros, pero en distinto orden

```python
lista = [['1','alfonso','IT',45000],
         ['3','alfredo','CS',85000],
         ['2','maria','CS',32000]
        ]

columnas = ['ID','NOMBRE','AREA','SALARIO']
df_1 = spark.createDataFrame(lista,columnas)
df_1.printSchema()
df_1.show(truncate=False)
```

```python
lista = [['1','alfonso','IT',45000],
         ['2','alfredo','CS',85000],
         ['3','maria','CS',32000]
        ]

columnas = ['ID','NOMBRE','AREA','SALARIO']
df_2 = spark.createDataFrame(lista,columnas)
df_2.printSchema()
df_2.show(truncate=False)
```

```python
#from pyspark.sql import DataFrame
import pyspark.sql.functions as F

def hash_df(df):
    
    df = df.withColumn('hash_value', F.hash(*sorted(df.columns))).select('hash_value')

    value = df.agg(F.sum('hash_value')).collect()[0][0]

    return value

df_1 = hash_df(df_1)
df_2 = hash_df(df_2)
assert df_1 == df_2
```
