#### Diferentes modos de escribir un Dataframe en un archivo

Se pueden especificar diferentes modos mientras se escribe el data frame en los archivos. 

- **df** = es el dataframe 
- **file_format** = puede ser cualquier formato válido como **text**, **csv**, **json**, **parquet**, **orc**.
- **mode(saveMode)** = puede ser mode(‘append’), mode(‘overwrite’), etc ...
- **path_to_folder** = ruta donde se guardará el archivo.

Distintas formas:

- df.write.mode(saveMode).file_format(path_to_folder)
- df.write.file_format(path_to_folder, mode=saveMode)
- df.write.mode(saveMode).format('file_format').save(path_to_folder)
- df.write.format('file_format').save(path_to_folder,mode=saveMode)

Opciones saveMode:

- **append**: Anexa el contenido del dataframe a los datos o tabla existentes.
- **overwrite** : Sobrescribe los datos existentes con el contenido del dataframe.
- **ignore** : Ignora la operación de escritura actual si los datos / la tabla ya existen sin lanzar ningún error.
- **error | errorifexists | default** : Lanza una excepción si los datos o la tabla ya existen.
