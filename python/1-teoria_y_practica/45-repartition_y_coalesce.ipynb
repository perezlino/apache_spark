### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Repartition y Coalesce

#### Repartition

Repartition se utiliza para el reshuffle de los datos en un número mayor o menor de particiones.

#### Coalesce 

Coalesce se utiliza normalmente para reducir (no permite aumentar) el número de particiones que hay que tratar como parte del procesamiento posterior
