### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Funciones de agregación

```
lista = [['1','alfonso','IT',45000],
         ['2','alfredo','CS',85000],
         ['3','maria','CS',32000],
         ['4','javiera','ECE',98000],
         ['5','tomas','IT',66000],
         ['6','jose','AKA',47000],
         ['7','cristian','PET',12000],
         ['8','nelson','AKA',33000],
        ]

columnas = ['ID','NOMBRE','AREA','SALARIO']
df = spark.createDataFrame(lista,columnas)
df.printSchema()
df.show(truncate=False)
```

#### Count

```
from pyspark.sql.functions import count

df.select(count("*")).show()
```

#### Sum

```
from pyspark.sql.functions import sum

df.select(sum("SALARIO")).show()
```

```
from pyspark.sql.functions import col

df.filter(col('AREA') == 'IT').select(sum("SALARIO")).withColumnRenamed("sum(SALARIO)","total_IT").show()
#df.filter(col('AREA') == 'IT').select(sum("SALARIO").alias("total_IT")).show()
```

```
#import pyspark.sql.functions as column ---> Lo mismo
#import pyspark.sql.functions as F ---> Otra forma
#df = df.agg(F.sum('SALARIO'))
from pyspark.sql import functions as column

df = df.agg(column.sum('SALARIO'))
df.show()
```

```
df.agg(column.sum('SALARIO')).collect()
```

```
df.agg(column.sum('SALARIO')).collect()[0]
```
