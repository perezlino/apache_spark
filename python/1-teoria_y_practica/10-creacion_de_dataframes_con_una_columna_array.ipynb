#### Forma 1

```python
array_appliance = [
        ('Raja',['TV','Refrigerator','Oven','AC']),
        ('Raghav',['AC','Washing machine',None]),
        ('Ram',['Grinder','TV']),
        ('Ramesh',['Refrigerator','TV',None]),
        ('Rajesh',None)]

df = spark.createDataFrame(data=array_appliance, schema = ['name','appliances'])
df.printSchema()
display(df)
```
