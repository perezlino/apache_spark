### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Date patterns

https://spark.apache.org/docs/latest/sql-ref-datetime-pattern.html

```
from pyspark.sql.functions import *

data = [['1','2020-02-01'],
        ['2','2019-03-01'],
        ['3','2021-06-15']
       ]

df = spark.createDataFrame(data,['id','input'])
```

```
df_modif = df.select(col('input'),to_date(col('input'),'yyyy-MM-dd').alias('fecha_nueva')).drop('input')

df_modif.printSchema()
df_modif.show()
```

```
df_modif.withColumn('dia_abreviado', date_format('fecha_nueva','E')).show(truncate=False)
```

```
df_modif.withColumn('dia_abreviado', date_format('fecha_nueva','EEEE')).show(truncate=False)
```
