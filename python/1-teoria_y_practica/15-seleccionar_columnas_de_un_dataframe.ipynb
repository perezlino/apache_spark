### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Seleccionar columnas de un dataframe

```
df = spark.read.option("inferSchema","true") \
               .option("header","true") \
               .csv("/FileStore/tables/raw/circuits.csv")
df.show(n=3, truncate=False)
```

##### Forma 1

```
df.select(['circuitId','circuitRef']).show()
```

##### Forma 2

```
df.select('circuitId','circuitRef').show()
```

##### Forma 3

```
from pyspark.sql.functions import col

df.select(col('circuitId'),col('circuitRef')).show()
```

##### Forma 4

```
df.select(df['circuitId'],df['circuitRef']).show()
```

##### Forma 5

```
df.select(df.circuitId, df.circuitRef).show()
```
