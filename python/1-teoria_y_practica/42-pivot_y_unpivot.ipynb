### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Pivot y Unpivot

```
data = [('ABC','Q1',2000),
        ('ABC','Q2',3000),
        ('ABC','Q3',1000),
        ('ABC','Q4',4000),
        ('XYZ','Q1',6000),
        ('XYZ','Q2',7000),
        ('XYZ','Q3',9000),
        ('XYZ','Q4',8000),
        ('KLM','Q1',9000),
        ('KLM','Q2',2000),
        ('KLM','Q3',5000),
        ('KLM','Q4',7000),
       ]

columnas = ['compania','trimestre','ganancia']

df = spark.createDataFrame(data, columnas)

df.printSchema()
df.show()
```

#### Pivot

```
df_pivot = df.groupBy('compania').pivot('trimestre').sum('ganancia')

df_pivot.show()
```

#### Unpivot

```
df_unpivot = df_pivot.selectExpr("compania","stack(4,'Q1',Q1,'Q2',Q2,'Q3',Q3,'Q4',Q4) as (trimestre,ganancia)")

df_unpivot.show()
```
