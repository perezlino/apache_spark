### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Extraer la primera y última fila de un Dataframe

```
data = [['1','sravan','company 1'],
        ['2','ojaswi','company 2'],
        ['3','bobby','company 3'],
        ['4','rohith','company 4'],
        ['5', 'alfonso','company 5']
       ]

columnas = ['id_empleado','nombre','empresa']

df = spark.createDataFrame(data, columnas)

df.show(truncate=False)
```

##### Ejemplo 1

```
df.select('empresa').distinct().collect()
```

```
for empresa_lista in df.select('empresa').distinct().collect():
  print(empresa_lista)
```

```
for empresa_lista in df.select('empresa').distinct().collect():
  print(empresa_lista.empresa)
```
