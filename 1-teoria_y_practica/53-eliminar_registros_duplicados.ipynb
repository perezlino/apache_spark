### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Eliminar registros duplicados

**distinct( )** , **dropDuplicates( )** y **drop_duplicates()** cumplen la misma funci√≥n.

```
from pyspark.sql.functions import *

estudiante = [(1,"Alfonso","Science",86,"A",90),
              (1,"Alfonso","Science",86,"A",90),
              (2,"Maria","Math",56,"R",55),
              (3,"Javiera","English",77,"A",85),
              (3,"Javiera","English",77,"A",85),
              (4,"Andres","Science",89,"A",98),
              (5,"Tomas","Math",69,"A",70),
              (5,"Tomas","Math",69,"A",70),
              (6,"Raul",'Phisics',45,"R",50),
              (7,"Pedro","English",61,"A",70)
             ]
schema = ["id","nombre","ramo","puntaje","status","asistencia"]

df = spark.createDataFrame(data=estudiante, schema=schema)
df.printSchema()
df.show()
```

#### dropDuplicates()

```
df.dropDuplicates().show()
```

#### drop_duplicates()

```
df.drop_duplicates().show()
```

```
df.drop_duplicates(subset=None).show()
```

#### distinct()

```
df.distinct().show()
```

#### Eliminar duplicados basado en una columna(s) especifica(s)

```
df.dropDuplicates('id').show()
```

```
df.dropDuplicates(['id']).show()
```

```
df.dropDuplicates(subset = ['id']).show()
```

```
df.dropDuplicates(['id','ramo']).show()
```

```
df.drop_duplicates(['id','ramo']).show()
```
