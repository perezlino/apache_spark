### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Obtener la potencia de una columna

```
from pyspark.sql.functions import col, lit, regexp_replace

data = [(1,'nombre A','Gibson',5,'5'),
        (2,'nombre B','Fender',3,'3'),
        (3,'nombre C','Ibanez',2,'2')
       ]

df = spark.createDataFrame(data).toDF('id','nombre','empresa','cantidad','cant_string')
df.printSchema()
df.show(truncate=False)
```

#### Se debe anteponer siempre la función “col”

```
df.withColumn('potencia',col('cantidad')**2).show(truncate=False)
```

#### Se busca elevar al cuadrado una columna del tipo String

```
df.withColumn('potencia',col('cant_string')**2).show(truncate=False)
```

#### pow()

```
df.withColumn('potencia',pow(col('cantidad'),lit(2))).show(truncate=False)
```

```
df.withColumn('potencia',pow(col('cantidad'),2)).show(truncate=False)
```

##### También sirve para columnas del tipo String

```
df.withColumn('potencia',pow(col('cant_string'),lit(2))).show(truncate=False)
```
