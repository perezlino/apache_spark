### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Revisar resumen estadístico de columnas de tipo numérico

Otra tarea común es calcular las estadísticas de resumen de una columna o conjunto de columnas. Podemos utilizar el método describe para lograr exactamente esto. Esto tomará todas las columnas numéricas y calculará el recuento, la media, la desviación estándar, el mínimo y el máximo. Deberías usar esto principalmente para verlo en la consola porque el esquema podría cambiar en el futuro:

#### describe( )

```
df = spark.read.format("csv").\
     option("inferSchema",True).\
     option("header",True).\
     option("sep",",").\
     load(['/FileStore/tables/by-day/2010_12*.csv'])

df.show(n=3, truncate=False, vertical=False)
```

```
df.describe("Quantity","UnitPrice").show(truncate=False)
```

#### summary( )

```
df.summary().show()
```

```
df.summary("count","min","25%","75%","max").show()
```

##### Para hacer un resumen de columnas específicas, primero selecciónelas:

```
df.select("InvoiceNo").summary("count").show()
```
