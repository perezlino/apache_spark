### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Cambiar a un tipo de dato unico para todas las columnas

```
from pyspark.sql.functions import *
from pyspark.sql.types import *

estudiante = [("Alfonso","Science",86,"A",90),
              ("Maria","Math",56,"R",None),
              ("Javiera","English",77,"A",85),
              ("Andres","Science",None,"A",None),
              ("Tomas","Math",69,"A",70),
              ("Raul",None,45,"R",50),
              ("Pedro","English",None,None,55)
             ]
schema = ["nombre","ramo","puntaje","status","asistencia"]

df.printSchema()
df = spark.createDataFrame(data=estudiante, schema=schema)
df.show()
```

```
schema_modif = df.select([col(c).cast(StringType()).alias(c) for c in df.columns])
schema_modif.printSchema()
schema_modif.show()
```
