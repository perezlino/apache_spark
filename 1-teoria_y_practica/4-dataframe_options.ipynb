### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Dataframe options

Múltiples atributos que se pueden utilizar dentro de la función **option/options** para definir cómo debe comportarse la operación de lectura y cómo debe interpretarse el contenido de la fuente de datos. La mayoría de los atributos enumerados a continuación pueden utilizarse en cualquiera de las funciones. Los atributos se pasan como string en la función option() pero no en la función options(). Esta diferencia se aclarará en la sección de ejemplos.


- **option()**: Esta función sólo admite un atributo/operación, pero la función option() puede utilizarse en serie.  

- **options()**: Esta función puede soportar múltiples atributos/operaciones usando pares de valores clave separados por comas.
