### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) Compresi贸n de escritura de archivos CSV

```
libros = [
          {'libro_id':1, 
           'nombre':'Python book', 
           'paginas':250,
           'stock':10
          },
          {'libro_id':2, 
           'nombre':'Java book', 
           'paginas':550,
           'stock':5
          },
          {'libro_id':3, 
           'nombre':'Scala book', 
           'paginas':350,
           'stock':8
          },
          {'libro_id':4, 
           'nombre':'C++ book', 
           'paginas':750,
           'stock':2
          },
          ]

from pyspark.sql import Row
df = spark.createDataFrame([Row(**libro) for libro in libros])
```

```
type(libros)
```

```
df.dtypes
```

```
display(df)
```

#### Forma 1 (normal)
##### El archivo se particiona en 8 partes

```
df.write.format('csv').save('/FileStore/tables/libros/', mode='overwrite', header=True)
```

#### Forma 2 (compresi贸n 'coalesce')
##### El archivo se particiona en 1 parte

```
df.coalesce(1).write.csv('/FileStore/tables/libros/', mode='overwrite', header=True)
```

#### Forma 3 (compresi贸n 'gzip')

```
df.coalesce(1).write.csv('/FileStore/tables/libros/', mode='overwrite', compression='gzip', header=True)
```

#### Forma 4 (compresi贸n 'snappy')

```
df.coalesce(1).write.csv('/FileStore/tables/libros/', mode='overwrite', compression='snappy', header=True)
```
