#### Forma 1

```
map_brand = [
        ('Raja',{'TV':'LG','Refrigerator':'Samsung','Oven':'Phillips','AC':'Voltas'}),
        ('Raghav',{'AC':'Samsung','Washing machine':'LG'}),
        ('Ram',{'Grinder':'Preethi','TV':''}),
        ('Ramesh',{'Refrigerator':'LG','TV':'Croma'}),
        ('Rajesh',None)]

df = spark.createDataFrame(data=map_brand, schema = ['name','brand'])
df.printSchema()
display(df)
```

#### Forma 2

```
from pyspark.sql.functions import element_at, lit

array = [({"a":1.0, "b":2.0},), ({"c":3.0, "d":4.0},)]

df = spark.createDataFrame(data=array, schema=['data'])
df.printSchema()
df.show()
```
