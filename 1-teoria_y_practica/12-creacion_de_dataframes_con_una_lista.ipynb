#### Utilizando una lista de strings

```
from pyspark.sql.types import StringType

auto = ['toyota','mazda','tesla','audi','ferrari']

df = spark.createDataFrame(auto, StringType()).withColumnRenamed('value','marca')
# df = spark.createDataFrame(auto, 'string')

df.printSchema()
df.show()
```

#### Utilizando una lista de enteros

```
from pyspark.sql.types import IntegerType

numero = [1,2,3,4,5,6]

df = spark.createDataFrame(numero, IntegerType()).withColumnRenamed('value','numero')
# df = spark.createDataFrame(auto, 'string')

df.printSchema()
df.show()
```

#### Utilizando una lista de tuplas

```
tupla = [(10,'diez'),(20,'veinte'),(30,'treinta')]

df = spark.createDataFrame(tupla, 'id int, nombre string')

df.printSchema()
df.show()
```
